```python
from transformers import pipeline
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
emotion_detector=pipeline("image-classification",model="trpakov/vit-face-expression")
```

```python
url = "https://images.pexels.com/photos/1239291/pexels-photo-1239291.jpeg"
response = requests.get(url)
image = Image.open(BytesIO(response.content)).convert("RGB")
image
```

```python
results = emotion_detector(image)

for result in results:
    label = result['label']
    score = round(result['score'] * 100, 2)
    print(f"{label}: {score}%")
```

```python
# print(results)
```

```python
top_emotion = max(results, key=lambda x: x['score'])
print(f"Emotion: {top_emotion['label']}")
print(f"Confidence: {round(top_emotion['score'] * 100, 2)}%")
```

```python
def detect_emotion(image):
    results = emotion_detector(image)
    top_emotion = max(results, key=lambda x: x['score'])
    label = top_emotion['label']
    score = round(top_emotion['score'] * 100, 2)
    return label, score
```

```python
label, score = detect_emotion(image)

print("\n--- RESULTS ---")
print(f"Emotion: {label}")
print(f"Confidence: {score}%")
```

```python
def show_emotion(image, label, score):
    plt.figure(figsize=(6, 6))
    plt.imshow(image)
    plt.title(f"Emotion: {label} ({score}%)",
              fontsize=16, color='green', pad=10)
    plt.axis('off')
    plt.show()

show_emotion(image, label, score)
```
