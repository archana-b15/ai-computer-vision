```python
from transformers import pipeline
from PIL import Image
import requests
from io import BytesIO
```

```python
captioner=pipeline("image-to-text",model="Salesforce/blip-image-captioning-base")
```

```python
url = "https://images.pexels.com/photos/1239291/pexels-photo-1239291.jpeg"
response = requests.get(url)
image = Image.open(BytesIO(response.content)).convert("RGB")
image
```

```python
result = captioner(image)
print(result)
```

```python
result = captioner(image, max_new_tokens=50)
caption = result[0]['generated_text']
print(f"Caption: {caption}")
```

```python
def generate_caption(image):
    result = captioner(image, max_new_tokens=50)
    caption = result[0]['generated_text']
    return caption
```

```python
import matplotlib.pyplot as plt
```

```python
def show_caption(image, caption):
    plt.figure(figsize=(8, 6))
    plt.imshow(image)
    plt.title(caption, fontsize=14,
              color='green', pad=10, wrap=True)
    plt.axis('off')
    plt.show()

caption = generate_caption(image)
show_caption(image, caption)
```

```python
url = "https://images.pexels.com/photos/346529/pexels-photo-346529.jpeg"
response = requests.get(url)
image = Image.open(BytesIO(response.content)).convert("RGB")
caption = generate_caption(image)
show_caption(image, caption)
```

```python
url = "https://images.pexels.com/photos/46798/the-ball-stadion-football-the-pitch-46798.jpeg"
response = requests.get(url)
image = Image.open(BytesIO(response.content)).convert("RGB")
caption = generate_caption(image)
show_caption(image, caption)
```
